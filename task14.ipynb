```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import random
from typing import List
```

### Гипотеза о равенстве средних против альтернатив a > b

```python
group1 = np.array([-1.11, -6.10, 2.42])
group2 = np.array([-2.29, -2.91])

num_samples = 5000
confidence = 0.95

observed_diff = np.mean(group1) - np.mean(group2)

bootstrap_diffs = []

for _ in range(num_samples):
    resampled_group1 = np.random.choice(group1, size=len(group1))
    resampled_group2 = np.random.choice(group2, size=len(group2))
    bootstrap_diffs.append(np.mean(resampled_group1) -
                           np.mean(resampled_group2) - observed_diff)

sorted_diffs = sorted(bootstrap_diffs)
critical_value = sorted_diffs[int((1 - confidence) * num_samples)]

upper_limit = observed_diff - critical_value

print(f"Верхняя граница доверительного интервала: {upper_limit:.4f}")
print(f"Оценка разницы средних: {observed_diff:.4f}")
```

### График

```python
std_deviation = np.sqrt(3)
threshold = 4.153


def power_function(theta):
    return 1 - norm.cdf(threshold, loc=theta, scale=std_deviation)


theta_range = np.linspace(-3, 11, 200)
power_curve = [power_function(theta) for theta in theta_range]

plt.figure(figsize=(8, 6))
plt.plot(theta_range, power_curve, color='navy')
plt.title("W(θ)")
plt.grid(True, alpha=0.3)
plt.xlim(-3, 11)
plt.ylim(-0.5, 1.5)
plt.show()
```
